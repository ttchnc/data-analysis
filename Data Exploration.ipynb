# Dataset First Look

**The file athlete_events.csv contains 271116 rows and 15 columns. Each row corresponds to an individual athlete competing in an individual Olympic event (athlete-events). The columns are:**

- **ID** - Unique number for each athlete
- **Name** - Athlete's name
- **Sex** - M or F
- **Age** - Integer
- **Height** - In centimeters
- **Weight** - In kilograms
- **Team** - Team name
- **NOC** - National Olympic Committee 3-letter code
- **Games** - Year and season
- **Year** - Integer
- **Season** - Summer or Winter
- **City** - Host city
- **Sport** - Sport
- **Event** - Event
- **Medal** - Gold, Silver, Bronze, or NA

# Data Exploration - Univariate

```python
def plot_distribution(df, list_col_names, cols=5, width=20, height=15, hspace=0.2, wspace=0.5):
    plt.style.use('seaborn-whitegrid')
    fig = plt.figure(figsize=(width,height))
    fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=wspace, hspace=hspace)
    rows = math.ceil(len(list_col_names) / cols)
    
    sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
    
    for i, column in enumerate(list_col_names):
        df_plot = df[df[column].isnull() == False].copy()
        
        ax = fig.add_subplot(rows, cols, i + 1)
        
        if df_plot.dtypes[column] == np.object:
            value_counts = df_plot[column].value_counts()
             
            if len(value_counts) < 11:
                g = sns.countplot(y=column, data=df_plot, palette='Set2')
                substrings = [s.get_text()[:18] for s in g.get_yticklabels()]
                g.set(yticklabels=substrings)
                plt.xticks(rotation=25)
            else:
                value_counts = value_counts.iloc[:5].sort_values(ascending=True)
                value_counts.plot.barh(ax=ax, alpha=0.5, color='c', legend=False) 
                ax.set_xlabel('Count')
            ax.set_title(column)
        else:
            g = sns.distplot(df_plot[column])
            plt.xticks(rotation=25)
    
```

```python
# deleting 1906 Olimpic games records from the dataset
df_data_unchanged = df_data.copy()
df_data = df_data[df_data['Year'] != 1906]
len(df_data_unchanged), len(df_data)
```

```python
print("First Summer Olimpic games were held in ", df_data[df_data['Season'] == 'Summer']['Year'].min())
print("The most recent Olympics were held in ", df_data[df_data['Year'] == df_data['Year'].max()]['City'].values[0])
```

```python
df_1986 = df_data[df_data['Year'] == 1896]
events1986 = set(df_1986['Event'])
nations1986 = list(set(df_1986['Team']))
nations1986 = [x for x in  nations1986 if ('/' not in x) & ('-' not in x)]
gender1986 = list(set(df_1986['Sex']))

if (len(gender1986) == 1) & (gender1986[0] == 'M'):
    str1 = str(len(df_1986)) + ' male competitors '
if (len(gender1986) == 1) & (gender1986[0] == 'F'):
    str1 = str(len(df_1986)) + ' female competitors '
if (len(gender1986) == 2):
    str1 = str(len(df_1986[df_1986['Sex'] == 'M'])) + ' male competitors and ' +\
            str(len(df_1986[df_1986['Sex'] == 'F'])) + ' female competitors '

df_2016 = df_data[df_data['Year'] == 2016]
events2016 = set(df_2016['Event'])
nations2016 = list(set(df_2016['Team']))
nations2016 = [x for x in  nations2016 if ('/' not in x) & ('-' not in x)]
gender2016 = list(set(df_2016['Sex']))

if (len(gender2016) == 1) & (gender2016[0] == 'M'):
    str2 = str(len(df_2016)) + ' male competitors '
if (len(gender2016) == 1) & (gender2016[0] == 'F'):
    str2 = str(len(df_2016)) + ' female competitors '
if (len(gender2016) == 2):
    str2 = str(len(df_2016[df_2016['Sex'] == 'M'])) + ' male competitors and ' +\
            str(len(df_2016[df_2016['Sex'] == 'F'])) + ' female competitors '  
    
print('The Olympics have increased in scope from a ', len(events1986), '-event competition with ' + str1 + 'from',
     len(nations1986), 'nations in 1986,' )
print('to ', len(events2016), ' events with ' + str2 + 'from ',
     len(nations2016), ' nations in 2016')
```
